```python
numerical_cols = annotsv_full.select_dtypes(include=np.number).columns.tolist()
atribut_b_split = annotsv_full[annotsv_full.clinsig == 'Benign']
atribut_p_split = annotsv_full[annotsv_full.clinsig == 'Pathogenic']

df_atrs = pd.DataFrame(columns=['Attribute_name', 'p_value'])

for col in numerical_cols:
    col = str(col)
    atribut_benign_split = atribut_b_split[col]
    atribut_pathogenic_split = atribut_p_split[col]
    
    attribute_pvalue = stats.mannwhitneyu(atribut_benign_split, atribut_pathogenic_split).pvalue
    
    df_atr = pd.DataFrame([[col, attribute_pvalue]], columns=['Attribute_name', 'p_value'])
    df_atrs = df_atrs.append(df_atr, ignore_index=True)

    if attribute_pvalue <= 0.001:
#         print(col, attribute_pvalue)
        plt.figure(figsize=(10, 8))
        sns.boxplot(data = annotsv_full, x = "clinsig", y = col, hue = "SV_type")
        fto_savefig = './results_isv_annotsv/comparison_attributes_annotsv_full/'+col+'.jpg'
        plt.savefig(fto_savefig, dpi=300, format='jpg', bbox_inches = 'tight', facecolor='white')
```
