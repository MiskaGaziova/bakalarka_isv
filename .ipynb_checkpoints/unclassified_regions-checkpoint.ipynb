```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from collections import Counter
import  math

from sklearn.feature_selection import RFE
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn import metrics
from matplotlib import rcParams
from matplotlib import pyplot
from xgboost import plot_importance
from numpy import sort
from numpy import loadtxt
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import RFE
%matplotlib inline
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
cnv_types = ['loss', 'gain']
cnv_type = cnv_types[1]

x_train_f = './data_annotsv/train_val_test_sets/train_val_test/x_train_' + cnv_type + '.tsv'
y_train_f = './data_annotsv/train_val_test_sets/train_val_test/y_train_' + cnv_type + '.tsv'
x_val_f = './data_annotsv/train_val_test_sets/train_val_test/x_val_'+ cnv_type + '.tsv'
y_val_f = './data_annotsv/train_val_test_sets/train_val_test/y_val_'+ cnv_type + '.tsv'
x_test_f = './data_annotsv/train_val_test_sets/train_val_test/x_test_' + cnv_type + '.tsv'
y_test_f = './data_annotsv/train_val_test_sets/train_val_test/y_test_'+ cnv_type + '.tsv'

x_train = pd.read_csv(x_train_f, sep='\t')
y_train = pd.read_csv(y_train_f, sep='\t')
x_val = pd.read_csv(x_val_f, sep='\t')
y_val = pd.read_csv(y_val_f, sep='\t')
x_test = pd.read_csv(x_test_f, sep='\t')
y_test = pd.read_csv(y_test_f, sep='\t')

ATTRIBUTES_TO_USE = [
    'morbidGene_count', 
    'min_D_Frequency_3',
    'ExAC_zScore_indel',
    'ExAC_zScore_pLI',
    'HI_DDDperc_haploinsufficient_gene_count',
    'CGscore_indel_gene_count_2', 
    'gene_count',
    'length'
]

if cnv_type == 'gain':
    ATTRIBUTES_TO_USE.remove('HI_DDDperc_haploinsufficient_gene_count')
    ATTRIBUTES_TO_USE.remove('CGscore_indel_gene_count_2')

x_train = x_train[ATTRIBUTES_TO_USE]
x_val = x_val[ATTRIBUTES_TO_USE]
x_test = x_test[ATTRIBUTES_TO_USE]

y_train = y_train.clinsig_bin
y_val = y_val.clinsig_bin
y_test = y_test.clinsig_bin
```

```python
# nacitaj natrenovany model
model_f = './results_isv_annotsv/models/xgb_'+ cnv_type + '.plk'
xgb_clf = pickle.load(open(model_f, 'rb'))
```

```python
TN, FP, FN, TP = metrics.confusion_matrix(y_val, xgb_clf.predict(x_val)).ravel()
accuracy = ((TP + TN)/(TP + TN + FP + FN))*100
accuracy
```

```python
# pouzi (na vytvorenie grafcoka, aby sme zistili, kde pada najviac falosnych predikcii)

validation_set = pd.concat([pd.DataFrame(list(y_val.clinsig_bin), columns=['clinvar_clinsig']), pd.DataFrame(xgb_clf.predict(x_val), columns=['Predicted']), pd.DataFrame(xgb_clf.predict_proba(x_val), columns=['Benign', 'Pathogenic'])], axis=1, ignore_index=False)
validation_set = validation_set.sort_values(by='Pathogenic')
validation_set = validation_set.reset_index()
```

```python
accuracy_all = []
precision_all = []
recall_all = []
specificity_all = []
startPathogenic = []
endPathogenic = []
unclassified_wrong_predictions_count_all = [] # kolko chybnych na jednotlivych usekoch (kolko realne predikovane chybne)
pathogenic_wrong_predictions_count_all = [] # kolko chybnych na useku predikovanych ako patogenne (kolko na useku patogennych je benignych)
BIN_SIZE = int(validation_set.shape[0] * 0.05) # posuvanie sa po piatich percentach

for i in range(1, (validation_set.shape[0] - BIN_SIZE)): # pre 5% (163): 3110): pre 20% (654) 2619 pre 15% (490) 2783 pre 10% (327) 2946; pre 2.5% (81) 3192
    unclassified = validation_set.iloc[i:i+BIN_SIZE] # bez tohoto accuracy, precision, recall
    
    benign = validation_set[:i][['clinvar_clinsig']]
    pathogenic = validation_set[i + BIN_SIZE:][['clinvar_clinsig']]
    reclassified = pd.concat([pathogenic, benign])
    unclassified = validation_set.iloc[i:i+BIN_SIZE] # bez tohoto accuracy, precision, recall
    benign = validation_set[:i][['clinvar_clinsig']]
    pathogenic = validation_set[i + BIN_SIZE:][['clinvar_clinsig']]
    startPathogenic.append(validation_set.loc[i, 'Pathogenic'])
    endPathogenic.append(validation_set.loc[i+BIN_SIZE, 'Pathogenic'])
    unclassified_wrong_predictions = (unclassified.Predicted != unclassified.clinvar_clinsig).sum()
    unclassified_wrong_predictions_count_all.append(unclassified_wrong_predictions)
#     pathogenic_wrong_predictions = (pathogenic.Predicted != pathogenic.clinvar).sum()
#     pathogenic_wrong_predictions_count_all.append(pathogenic_wrong_predictions)
    
    
    tp = len(pathogenic[pathogenic.clinvar_clinsig == 1])
    fp = len(pathogenic[pathogenic.clinvar_clinsig == 0])
    tn = len(benign[benign.clinvar_clinsig == 0])
    fn = len(benign[benign.clinvar_clinsig == 1])
    recall = tp / (tp + fn)
    precision = tp / (tp + fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    specificity = (tn / (tn + fp))
    accuracy_all.append(accuracy)
    precision_all.append(precision)
    recall_all.append(recall)
    specificity_all.append(specificity)
    
```

```python
# (setup tabulky) Vykreslenie grafov
threshold_pathogenic_benign = pd.concat([pd.DataFrame(unclassified_wrong_predictions_count_all), pd.DataFrame(startPathogenic),  pd.DataFrame(endPathogenic), pd.DataFrame(accuracy_all), pd.DataFrame(precision_all), pd.DataFrame(recall_all), pd.DataFrame(specificity_all)],  axis = 1) #  pd.DataFrame(logreg_sorted.Benign), pd.DataFrame(logreg_sorted.Pathogenic
threshold_pathogenic_benign_columns_names = ['Counts of wrong predictions in unclassified region', 'PathogenicStart', 'PathogenicEnd', 'Accuracy', 'Precision', 'Recall', 'Specificity']  # 'Benign_probability', 'Pathogenic_probability , 'Probability of pathogenic prediction (start)'
threshold_pathogenic_benign.columns = threshold_pathogenic_benign_columns_names
threshold_pathogenic_benign['Index'] = threshold_pathogenic_benign.index
threshold_pathogenic_benign
```

```python
threshold_pathogenic_benign_f = './results_isv_annotsv/unclassified_tresholds/thresholds_pathogenic_benign_5perc_'+cnv_type+'.tsv'
threshold_pathogenic_benign.to_csv(threshold_pathogenic_benign_f, sep='\t', index=False)
```

```python
# Hladanie useku, s najvacsim poctom chybnych predikcii
maxCount_wrong_predictions = max(threshold_pathogenic_benign['Counts of wrong predictions in unclassified region'])
wrong_pred = np.where(threshold_pathogenic_benign['Counts of wrong predictions in unclassified region'] >= maxCount_wrong_predictions)
threshold_pathogenic_benign.loc[wrong_pred]
```

```python
# Hladanie useku, s najvyssou  presnosnostou
maxAccuracy = max(threshold_pathogenic_benign['Accuracy'])
maxAc = np.where(threshold_pathogenic_benign['Accuracy'] >= maxAccuracy)
threshold_pathogenic_benign.loc[maxAc]
```
