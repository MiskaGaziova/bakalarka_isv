```python
# pouzi (na vytvorenie grafcoka, , kde pada najviac falosnych predikcii)

validation_set = pd.concat([pd.DataFrame(list(y_val.clinsig_bin), columns=['clinvar_clinsig']), pd.DataFrame(xgb_clf.predict(x_val), columns=['Predicted']), pd.DataFrame(xgb_clf.predict_proba(x_val), columns=['Benign', 'Pathogenic'])], axis=1, ignore_index=False)
validation_set = validation_set.sort_values(by='Pathogenic')
validation_set = validation_set.reset_index()
```

```python
def probability_of_prediction(row):
    if row.clinvar_clinsig == 0:
        if row.Predicted == 1:
            category = 'FP'
        else:
            category = 'TN'
    else:
        if row.Predicted == 1:
            category = 'TP'
        else:
            category = 'FN'
    return category

# validation_set['Category'] = validation_set.apply(probability_of_prediction, axis=1)
# test_predictions['Category'] = test_predictions.apply(probability_of_prediction, axis=1)
# loss['Category'] = loss.apply(probability_of_prediction, axis=1)
# gain['Category'] = gain.apply(probability_of_prediction, axis=1)
```

```python
TN, FP, FN, TP = metrics.confusion_matrix(validation_set.clinvar_clinsig, validation_set.Predicted).ravel()
TN, FP, FN, TP
```

# Vyhodnotenie na testovacom sete

```python
cnv_types = ['loss', 'gain']
for cnv_type in cnv_types:

    x_train_f = './data_annotsv/train_val_test_sets/train_val_test/x_train_' + cnv_type + '.tsv'
    y_train_f = './data_annotsv/train_val_test_sets/train_val_test/y_train_' + cnv_type + '.tsv'
    x_val_f = './data_annotsv/train_val_test_sets/train_val_test/x_val_'+ cnv_type + '.tsv'
    y_val_f = './data_annotsv/train_val_test_sets/train_val_test/y_val_'+ cnv_type + '.tsv'
    x_test_f = './data_annotsv/train_val_test_sets/train_val_test/x_test_' + cnv_type + '.tsv'
    y_test_f = './data_annotsv/train_val_test_sets/train_val_test/y_test_'+ cnv_type + '.tsv'

    x_train = pd.read_csv(x_train_f, sep='\t')
    y_train = pd.read_csv(y_train_f, sep='\t')
    x_val = pd.read_csv(x_val_f, sep='\t')
    y_val = pd.read_csv(y_val_f, sep='\t')
    x_test = pd.read_csv(x_test_f, sep='\t')
    y_test = pd.read_csv(y_test_f, sep='\t')

    ATTRIBUTES_TO_USE = [
        'morbidGene_count', 
        'min_D_Frequency_3',
        'ExAC_zScore_indel',
        'ExAC_zScore_pLI',
        'HI_DDDperc_haploinsufficient_gene_count',
        'CGscore_indel_gene_count_2', 
        'gene_count',
        'length'
    ]

    if cnv_type == 'gain':
        ATTRIBUTES_TO_USE.remove('HI_DDDperc_haploinsufficient_gene_count')
        ATTRIBUTES_TO_USE.remove('CGscore_indel_gene_count_2')

    x_train = x_train[ATTRIBUTES_TO_USE]
    x_val = x_val[ATTRIBUTES_TO_USE]
    x_test = x_test[ATTRIBUTES_TO_USE]


    # nacitaj natrenovany model
    model_f = './results_isv_annotsv/models/xgb_'+ cnv_type + '.plk'
    xgb_clf = pickle.load(open(model_f, 'rb'))

    test_set = pd.concat([pd.DataFrame(list(y_test.clinsig_bin), columns=['clinvar_clinsig']), pd.DataFrame(xgb_clf.predict(x_test), columns=['Predicted']), pd.DataFrame(xgb_clf.predict_proba(x_test), columns=['Benign', 'Pathogenic'])], axis=1, ignore_index=False)
    test_set = test_set.sort_values(by='Pathogenic')
    test_set = test_set.reset_index()
    test_set['cnv_type'] = cnv_type
    test_set['Category'] = test_set.apply(probability_of_prediction, axis=1)
    test_set['Category_5perc_uncl'] = test_set.apply(get_tn_fp_uncl5perc, axis=1)
    test_set['Category_10perc_uncl'] = test_set.apply(get_tn_fp_uncl10perc, axis=1)
    
    if cnv_type == 'loss':
        loss = test_set
    else:
        gain = test_set
        
test_predictions = pd.concat([loss, gain], ignore_index=True)
```

```python
# pre 10 percentnu stratu z validacneho 


def get_tn_fp_uncl10perc(row): 
    uncertain_down_gain = 0.04 # 0.34
    uncertain_up_gain = 0.94  # 0.82
    uncertain_down_loss = 0.11 # 0.14
    uncertain_up_loss = 0.93  #0.95

    if row.cnv_type == 'loss':
        uncertain_down_threshold = uncertain_down_loss
        uncertain_up_threshold = uncertain_up_loss
    else:
        uncertain_down_threshold = uncertain_down_gain
        uncertain_up_threshold = uncertain_up_gain
        
    if row.clinvar_clinsig == 0:
        if row.Pathogenic > uncertain_up_threshold:
            return 'FP'
        elif row.Pathogenic >= uncertain_down_threshold and row.Pathogenic <= uncertain_up_threshold:
            return 'Unclassified'
        else:
            return 'TN'
    else:
        if row.Pathogenic > uncertain_up_threshold:
            return 'TP'
        elif row.Pathogenic >= uncertain_down_threshold and row.Pathogenic <= uncertain_up_threshold:
            return 'Unclassified'
        else:
            return 'FN'

# loss['Category_new'] = loss.apply(get_tn_fp, axis = 1)
# gain['Category_new'] = gain.apply(get_tn_fp, axis = 1)
# test_predictions['Category_new'] = test_predictions.apply(get_tn_fp, axis = 1)

# train = sns.boxplot(data = prediction_sorted, x = "Category_new", y = 'Pathogenic', order = ['TN', 'FN', 'Unclassified','FP', 'TP'])
```

```python
cnv_types = ['loss', 'gain']
categories = ['Category', 'Category_5perc_uncl','Category_10perc_uncl']
# 'Accuracy', 'Accuracy_5%unclassified','Accuracy_10%unclassified'
all_acc = pd.DataFrame(columns=['cnv_type', 'Category', 'Category_5perc_uncl','Category_10perc_uncl'])

for cnv_type in cnv_types:
    data = test_predictions[test_predictions.cnv_type == cnv_type]
    x = pd.DataFrame([[cnv_type]], columns=['cnv_type'])
    for cat in categories:
        TrueFalse_category = pd.DataFrame.from_dict(Counter(data[cat]), orient='index').transpose()
        TN = TrueFalse_category.iloc[0] ['TN']
        FN = TrueFalse_category.iloc[0] ['FN']
        TP = TrueFalse_category.iloc[0] ['TP']
        FP = TrueFalse_category.iloc[0] ['FP']
        print(TN, FN, TP, FP)
        print('Accuracy:',cat, (TP + TN)/(TP + TN + FP + FN))
        accuracy = ((TP + TN)/(TP + TN + FP + FN))
        c = pd.DataFrame([[accuracy]], columns=[cat])
        x = pd.concat([x, c], axis=1, ignore_index=False)
    all_acc = all_acc.append(x)

all_acc
all_acc.columns = ['cnv', 'Accuracy', 'Accuracy_5%unclassified','Accuracy_10%unclassified']
all_acc.to_csv('./results_isv_annotsv/test_predictions/test_predictions_loss_gain_accuracy_uncl.tsv', sep='\t', index=False)
```
